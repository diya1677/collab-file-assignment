# **Project Summary -**

This project analyzes the United Nations Global Terrorism Analysis (UNGTA) dataset, which contains detailed records of terrorist incidents worldwide from 1970 to 2017. The primary objective is to identify patterns, trends, and impacts of terrorism using data analysis and visualization techniques.

The dataset includes information such as date, location (country, region, city), attack types, weapon types, target groups, and casualty figures. Using Python libraries like Pandas, NumPy, Matplotlib, and Seaborn, the data was cleaned, processed, and analyzed. Exploratory Data Analysis (EDA) was performed to understand global terrorism trends. Visualizations revealed that terrorist activities have increased significantly after the year 2000. Further analysis showed that bombing and explosion are the most common attack types, and explosives are the most frequently used weapons. Impact assessment using casualty data demonstrated that some periods experienced significantly higher casualties, indicating intense conflict phases.
The insights derived from this project can help governments, security agencies, and international organizations better understand terrorism patterns and make informed decisions. Overall, this project demonstrates how data analysis can be used to uncover meaningful insights and support strategic planning in addressing global security challenges.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Fixing invalid values first
df['imonth'] = df['imonth'].replace(0, 1)
df['iday'] = df['iday'].replace(0, 1)

# Rename columns for datetime
date_df = df[['iyear','imonth','iday']].rename(columns={
    'iyear': 'year',
    'imonth': 'month',
    'iday': 'day'
})

# Create date column
df['date'] = pd.to_datetime(date_df, errors='coerce')
````

````python
# Chart - 1 visualization code
# Group data by year and attack type
attack_trend = df.groupby(['iyear','attacktype1_txt']).size().reset_index(name='count')

# Plot
plt.figure(figsize=(14,7))
sns.lineplot(data=attack_trend, x='iyear', y='count', hue='attacktype1_txt')

plt.title("Year-wise Terrorist Attacks by Attack Type")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.legend(title="Attack Type", bbox_to_anchor=(1.05,1))
plt.show()
````
